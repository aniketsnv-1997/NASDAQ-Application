```python
#web scraping libraries
from bs4 import BeautifulSoup
from requests import get

#other supplementry libraries
from random import randint
from time import time,sleep
from IPython.core.display import clear_output

#general libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://www.nasdaq.com/screening/companies-by-industry.aspx?industry=Finance&region=Asia"
response = get(url)
```

```python
print(response.text) ##prints the html code
print(response.encoding) ##prints the character encoding used in the html code
print(response.headers)  ##prints the meta data (headers) associated with the url
print(response.status_code) ##returns the status code
```

```python
page_html = BeautifulSoup(response.text, 'html.parser') ##we have converted the response object to an object 
print(page_html) #of BeautifulSoup because find() can be applied only on BeautifulSoup class objects
```

```python
'''1. find(tag_name, [attrs={id:id_name}]) is a method in BeautifulSoup class which gets the html code of a
particular tag whose name is given as 1st arguement to the method. 

2. In case of presence of multiple same tags, we can specify the tag of our interest with the help of the 2nd
arguement - "attrs" which is a dictionary where
    key - desired identifier(id/class) of desired tag
    value - name of the selected identifier
'''

data_dictionary={}
table_data = page_html.find("table", attrs={'id':'CompanylistResults'})
print(table_data)
```

```python
'''Now we have to iterate through each <tr> tag because tr contains each row of table.
For this we will use find_all() with tr as 1st arguement because we have to find all code pieces with tr tag
find() would return only 1 code snippet

*Note: no 2nd arguement is given because all tr tags from contents var are desired
'''
tr_data = table_data.find_all('tr')
print("data of all rows in table")
print(tr_data, len(tr_data))
print('\n')
```

```python
##This code block contains code to carryout only for single values in list. 

print("data of 1st row(headings row) in table")
th_data = tr_data[0].find_all('th')
print(th_data)

print('\n\n')

print("data of other rows in table")
td_data = tr_data[1].find_all('td')
print(td_data)

for td in td_data:
    print(td.text.strip())

'''Now as we have demonstrated that we can get the desired data(column name), we will retrieve all the column
headings with the help of below code and will append the retrieved value to above defined dictionary - 
data_dictionary as individual keys with null values!
'''
for th in th_data:
    key =  th.text.strip()
    data_dictionary[key] = ''

print(data_dictionary)
```

```python
'''This code block contains the looped code for doing the above tasks. The actual scraping of data will not include
headings of the desired table! It will have only rows
'''
table_data_list = []
row_data_list = []

#tr is one/five value in tr_data list (total rows = 5 in tr_data, therefore loop will execute 5 times)
for tr in tr_data: 
    
    #td_data contains all <td> tags from current tr
    td_data = tr.find_all('td') 
    
    '''
    each <tr> contains either 1 or 7 <td> tags! But desired data is in those tr whose td count is 7!
    as td_data holds td's of each tr, we will select those <tr> with 7 <td>
    '''
    if len(td_data) == 7:
        
        #td is one/seven value in td_data list (total td = 7 in td_data, therefore loop will execute 7 times)
        for td in td_data: 
            
            #we do not want any empty <td>, therefore we filter empty <td> with the below if
            if td.text.strip() != '': 
                row_data_list.append(td.text.strip()) 

        #print("row data is {}\n".format(row_data_list))
        #after executing this loop 7 times, the row_list will have all data of 1 FULL ROW

        '''
        now we append this 1 row ready list to the the list holding the data of all rows of table i.e. 
        to the list holding the entire table data apart
        #Reminder: we have not included the column headings in this
        '''
        table_data_list.append(row_data_list)
        
        '''after appending the row data to table data, we empty the row_data_list so that it can store 
        next row data
        '''
        row_data_list = []
    
#each value in tabe_data_list represents 1 row from table. Thus 2 rows
print("Final table data is \n{}".format(table_data_list))
print("\nCount of rows in table is \n{}".format(len(table_data_list)))
```
